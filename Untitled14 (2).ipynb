```python
import torch
import torchvision
import torchvision.models as models
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from PIL import Image

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def preprocess_image(image_path):
    image = Image.open(image_path)
    image = transform(image)
    image = torch.unsqueeze(image, 0)
    return image

image_dir = 'images'
dataset = ImageFolder(root=image_dir, transform=transform)
data_loader = DataLoader(dataset, batch_size=1, shuffle=False)

vgg16 = torchvision.models.vgg16(pretrained=True)

vgg16.eval()
list_normal = []
for inputs, _ in data_loader:
    with torch.no_grad():
        outputs = vgg16(inputs)
        output_prob = torch.nn.functional.softmax(outputs[0], dim=0)
        class_names = dataset.classes
        if class_names[1] == 'normal_images':
          normal_prob = output_prob[1].item()
          inverted_prob = output_prob[0].item()
          print("Thacther dist:", abs(float(normal_prob-inverted_prob)/float(normal_prob+inverted_prob)))





'''
print("output_prob for Normal Image:")
print(normal_output_prob)

print("\noutput_prob for Inverted Image:")
print(inverted_output_prob)
'''


```
